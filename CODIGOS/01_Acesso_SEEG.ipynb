```python
import requests
import io
import pandas as pd
```

```python
# URL original do Google Drive do SEEG
original_url = "https://drive.google.com/file/d/1a2fXQwGZ7Qb7hlOGg5drowdfAiupSYhJ/view?usp=drive_link" #dados ar6 do estado do MT

# Extrai o ID do arquivo
file_id = original_url.split("/d/")[1].split("/")[0]
download_url = f"https://drive.google.com/uc?id={file_id}&export=download"

# Fazendo o download
response = requests.get(download_url)

# Lê o conteúdo do CSV diretamente na memória
csv_data = io.StringIO(response.text)
```

```python
df = pd.read_csv(csv_data)
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
# Extrair o nome do município e o estado usando expressões regulares
df['Municipio'] = df['Cidade'].str.extract(r'^(.*) \(')  # Extrai tudo antes de " ("
df['Estado'] = df['Cidade'].str.extract(r'\((.*)\)$')   # Extrai tudo entre parênteses

# Visualizar as primeiras linhas para verificar o resultado
print(df[['Cidade', 'Municipio', 'Estado']].head())

#Aqui, embora os dados sejam acessados por estado e todas os registos da coluna 'Estado' sejam iguais, este atributo será importante caso, em códigos futuros, seja feita a junção dos dados de vários estados
```

Compreensão da relação entre as variáveis categóricas


```python
# Seleciona as colunas categóricas (tipo 'object')
categorical_columns = df.select_dtypes(include=['object']).columns

# Exibe os valores únicos de cada coluna categórica
for column in categorical_columns:
    print(f"Valores únicos para a coluna '{column}':")
    print(df[column].unique())  # Exibe os valores únicos
    print("\n")  # Adiciona uma linha em branco entre as colunas
```

```python
# Agrupar e obter os valores únicos de 'Categoria emissora' para cada 'Setor de emissão'
grouped_01 = df.groupby('Setor de emissão')['Categoria emissora'].unique().reset_index()

# Expandir as listas de 'Categoria emissora' para linhas individuais
expanded_01= grouped_01.explode('Categoria emissora')
expanded_01
```

```python
#Verifica se há valores de categoria emissora que aparece em mais de um setor de emissão


# Agrupar por 'Categoria emissora' e listar os setores únicos para cada categoria
category_groups_01 = df.groupby('Categoria emissora')['Setor de emissão'].nunique()

# Filtrar as categorias que aparecem em mais de um setor
categories_in_multiple_sectors_01 = category_groups_01[category_groups_01 > 1]

categories_in_multiple_sectors_01
```

```python
# Agrupar por 'Categoria emissora' e listar os valores únicos de 'Sub-categoria emissora' para cada grupo
grouped_02 = df.groupby('Categoria emissora')['Sub-categoria emissora'].unique().reset_index()

# Expandir as listas de 'Categoria emissora' para linhas individuais
expanded_02= grouped_02.explode('Sub-categoria emissora')
expanded_02
```

```python
#Verifica se há valores de sub-categoria emissora que aparece em mais de uma categoria emissora
subcategory_groups = df.groupby('Sub-categoria emissora')['Categoria emissora'].nunique()

# Filtrar as subcategorias que aparecem em mais de uma categoria emissora
subcategories_in_multiple_categories = subcategory_groups[subcategory_groups > 1]

subcategories_in_multiple_categories
```

```python
# Agrupar por 'Sub-categoria emissora' e listar as categorias de emissora para as subcategorias que aparecem em mais de uma categoria
result = df[df['Sub-categoria emissora'].isin(subcategories_in_multiple_categories.index)] \
    .groupby('Sub-categoria emissora')['Categoria emissora'].unique() \
    .explode()  # Explodir as categorias para linhas individuais

result
```

```python
import plotly.graph_objects as go
import random
```

```python
# Preparar os dados para o Sankey
sankey_df = result.reset_index(name='Categoria emissora') \
    .rename(columns={'Sub-categoria emissora': 'source', 'Categoria emissora': 'target'})

# Adicionar uma coluna de valor (peso)
sankey_df['value'] = 1  # Aqui, cada conexão recebe peso 1; você pode ajustá-lo conforme necessário

# Criar lista única de nós
nodes = list(set(sankey_df['source']).union(set(sankey_df['target'])))
node_indices = {node: i for i, node in enumerate(nodes)}

# Converter os valores de origem e destino para índices
sankey_df['source_index'] = sankey_df['source'].map(node_indices)
sankey_df['target_index'] = sankey_df['target'].map(node_indices)

# Gerar cores claras aleatórias para os links
def generate_light_color():
    """Gera uma cor aleatória clara no formato hexadecimal."""
    r, g, b = [random.randint(150, 255) for _ in range(3)]  # Garantir valores altos para cores claras
    return f"#{r:02X}{g:02X}{b:02X}"

unique_sources = sankey_df['source'].unique()
colors = {source: generate_light_color() for source in unique_sources}
link_colors = sankey_df['source'].map(colors)

# Configurando os dados para o diagrama Sankey
sankey_data = go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes,
        color="lightgray"  # Cor dos nós
    ),
    link=dict(
        source=sankey_df['source_index'],
        target=sankey_df['target_index'],
        value=sankey_df['value'],
        color=link_colors  # Usar cores claras aleatórias para os links
    )
)

# Criando a figura
fig = go.Figure(sankey_data)

# Configurando o layout
fig.update_layout(
    title_text="Diagrama de Sankey - Subcategorias e Categorias Emissoras",
    font=dict(color="black", size=14)  # Configurar fonte global para preto
)

# Exibindo a figura
fig.show()
```
