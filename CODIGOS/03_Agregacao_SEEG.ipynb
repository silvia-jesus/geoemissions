```python
df = pd.read_csv('/content/drive/MyDrive/PESQUISAS/ENSAIO_MUDANCAS_CLIMATICAS/SAIDA/01_DADOS_SEEG/SEEG_MT.csv')
df.head()
```

```python
# Lista de anos para a agregação
anos = ['1997', '2003', '2010', '2015', '2023']

# Agrupar os dados por 'Municipio' e 'Setor de emissão'
df_grouped = df.groupby(['Setor de emissão','Municipio'])[anos].sum()

df_grouped
```

```python
output_dir = '/content/drive/MyDrive/PESQUISAS/ENSAIO_MUDANCAS_CLIMATICAS/SAIDA/03_AGREGACAO_SEEG'

# Agrupar os dados por 'Setor de emissão' e 'Municipio'
df_grouped = df.groupby(['Setor de emissão', 'Municipio'])[anos].sum()

# Iterar pelos valores únicos de 'Setor de emissão'
for setor in df['Setor de emissão'].unique():
    # Filtrar os dados para o setor atual
    df_setor = df_grouped.loc[setor]

    # Criar o nome do arquivo
    setor_filename = setor.replace('/', '_').replace(' ', '_')  # Evitar caracteres problemáticos
    filename = f"{output_dir}/{setor_filename}_municipal.csv"

    # Salvar o arquivo CSV
    df_setor.to_csv(filename)

    print(f'Arquivo salvo: {filename}')
```
