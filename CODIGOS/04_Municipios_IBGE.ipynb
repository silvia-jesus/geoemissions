```python
url = 'https://geoftp.ibge.gov.br/organizacao_do_territorio/malhas_territoriais/malhas_municipais/municipio_2022/Brasil/BR/BR_Municipios_2022.zip'
response = requests.get(url)
content = response.content
muni_BR = gpd.read_file(io.BytesIO(content), sep=',', compression='gzip', index_col=0, quotechar='"')
muni_BR.head()
```

```python
crs = muni_BR.crs
crs
```

```python
uf = 'MT'
```

```python
muni = muni_BR[muni_BR['SIGLA_UF'] == uf]
muni.shape
```

```python
muni_IBGE = muni['NM_MUN'].tolist()
```

```python
seeg = pd.read_csv('/content/drive/MyDrive/PESQUISAS/ENSAIO_MUDANCAS_CLIMATICAS/SAIDA/01_DADOS_SEEG/SEEG_MT.csv')
seeg.head()
```

```python
muni_SEEG = seeg['Municipio'].unique().tolist()
muni_SEEG
```

```python
comum = list(set(muni_IBGE) & set(muni_SEEG))  # Elementos comuns entre as listas
comum
```

```python
print(len(muni_IBGE))
print(len(muni_SEEG))
print(len(comum))
```

```python
# Convertendo as listas para conjuntos e fazendo a diferença
falta_no_IBGE = list(set(muni_SEEG) - set(muni_IBGE))
falta_no_IBGE
```

```python
# Crie uma cópia do DataFrame
muni_ = muni.copy()

# Renomeando a coluna na cópia
muni_.rename(columns={'NM_MUN': 'Municipio'}, inplace=True)
```

União de tabelas com os dados de emissão por setor nos anos selecionados

```python
files_path = '/content/drive/MyDrive/PESQUISAS/ENSAIO_MUDANCAS_CLIMATICAS/SAIDA/03_AGREGACAO_SEEG/*csv'
files_list = glob.glob(files_path)
print(len(files_list))
files_list
```

```python
# Loop para abrir cada arquivo e exibir o shape
for file in files_list:
    try:
        # Lê o arquivo CSV em um DataFrame
        df = pd.read_csv(file)

        # Exibe o nome do arquivo e o shape
        print(f'Arquivo: {file} | Shape: {df.shape}')
    except Exception as e:
        print(f'Erro ao processar o arquivo {file}: {e}')
```

```python
for file in files_list:
  df = pd.read_csv(file)
  setor = file.replace('/content/drive/MyDrive/PESQUISAS/ENSAIO_MUDANCAS_CLIMATICAS/SAIDA/03_AGREGACAO_SEEG/','').replace('_municipal.csv','')
  print(setor)
  gdf = muni_.merge(df, on='Municipio', how='left')
  gdf.to_file(f'/content/drive/MyDrive/PESQUISAS/ENSAIO_MUDANCAS_CLIMATICAS/SAIDA/04_MUNICIPIOS_IBGE/AGREGACAO_POR_SETOR/{setor}.shp')
  print(gdf.shape)
  print(gdf.crs)
```
