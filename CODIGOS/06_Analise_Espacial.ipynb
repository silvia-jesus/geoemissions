```python
# Loop pelos arquivos
for file in files_list:
    # Extrair o nome do setor
    setor = os.path.basename(file).replace('.shp', '').replace('_',' ')

    # Ler o shapefile
    gdf = gpd.read_file(file)

    # Testar normalidade para cada ano
    for ano in anos:
        if ano not in gdf.columns:
            print(f"A coluna '{ano}' não está presente no shapefile {setor}. Pulando...")
            continue

        # Filtrar valores não nulos
        valores = gdf[ano].dropna()

        # Realizar o teste de Shapiro-Wilk
        stat, p_value = shapiro(valores)

        # Determinar o resultado (normal ou não)
        normal = "Sim" if p_value > 0.05 else "Não"

        # Adicionar os resultados ao DataFrame
        resultados_normalidade.append({
            "Setor": setor,
            "Ano": ano,
            "Estatística": stat,
            "p-Valor": p_value,
            "Normal?": normal
        })
```

Transformação de Normalidade

```python
# Função para aplicar a melhor transformação
def melhor_transformacao(dados):
    """
    Aplica as transformações log, Box-Cox ou uma tentativa de transformação de potência
    para tentar normalizar os dados, dependendo da distribuição original.
    Retorna os dados transformados e o tipo de transformação utilizada.
    """
    # Verificar se os dados contêm valores negativos ou zero
    if (dados <= 0).any():
        # Tentar Box-Cox1p se houver valores zero ou negativos
        dados_transformados = boxcox1p(dados, 0)  # Corrigido: boxcox1p retorna apenas um valor
        tipo_transformacao = 'Box-Cox1p'
    else:
        # Aplicar transformação Box-Cox normal
        dados_transformados, _ = boxcox(dados)
        tipo_transformacao = 'Box-Cox'

    # Verificar normalidade dos dados transformados
    estatisticas, p_valor = normaltest(dados_transformados)

    # Verificar se os dados já são normais ou não
    if p_valor > 0.05:
        return dados_transformados, tipo_transformacao
    else:
        # Caso a Box-Cox não tenha sido suficiente, tentamos logaritmo
        dados_transformados = np.log(dados)
        tipo_transformacao = 'Log'
        return dados_transformados, tipo_transformacao
```

```python
def aplicar_transformacoes_gerais(gdf, anos_selecionados, setor):
    # Criar uma cópia do GeoDataFrame para não modificar o original
    gdf_transformado = gdf.copy()

    for ano in anos_selecionados:
        dados = gdf[ano].dropna()  # Remover valores N/A

        # Aplicar a transformação
        dados_transformados, tipo_transformacao = melhor_transformacao(dados)

        # Substituir a coluna do ano no GeoDataFrame com os dados transformados
        gdf_transformado[ano] = dados_transformados

        # Criar o histograma do dado transformado
        plt.figure(figsize=(10, 6))
        sns.histplot(dados_transformados, kde=True, bins=30, color='skyblue', stat='density')
        plt.title(f'{setor} - Histograma para o ano {ano} ({tipo_transformacao})')
        plt.xlabel(f'{ano} (Transformado)')
        plt.ylabel('Densidade')
        plt.show()

    # Salvar o GeoDataFrame transformado como um arquivo .shp
    output_file = f'/content/drive/MyDrive/PESQUISAS/ENSAIO_MUDANCAS_CLIMATICAS/SAIDA/06_ANALISE_ESPACIAL/SHP_TRANSFORMADO/{setor}_transformado.shp'
    gdf_transformado.to_file(output_file)
    print(f"Arquivo transformado salvo como: {output_file}")
```

```python
def calcular_moran_lisa(gdf, coluna, output_dir, nome_setor, ano):
    """
    Calcula o índice de Moran e LISA para uma coluna específica de um arquivo shapefile.
    """
    # Realiza o cálculo do Índice de Moran
    y = gdf[coluna].values
    w = Queen.from_dataframe(gdf)
    w.transform = 'r'
    moran = Moran(y, w)

    I = moran.I
    EI = moran.EI
    p = moran.p_norm

    print(f'******* {nome_setor} - {ano} *******')
    print(f'Índice de Moran: {I}')
    print(f'Valor esperado sob a hipótese de normalidade: {EI}')
    print(f'p-valor do I sob a hipótese de normalidade: {p}')

    if I < EI:
        print('Falha ao rejeitar H0: É uma distribuição aleatória')
    else:
        print('Rejeita H0: Não é uma distribuição aleatória')

    # Cálculo do Moran Local (LISA)
    moran_loc = Moran_Local(y, w)

    # Plot LISA
    fig, ax = plt.subplots(figsize=(10, 8))
    lisa_cluster(moran_loc, gdf, ax=ax, legend=False)

    # Adiciona título com o nome do setor e ano
    ax.set_title(f'Índice LISA - {nome_setor} ({ano})', fontsize=16)

    # # Ajusta a posição da legenda mais à direita
    # ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    # Salva o gráfico
    plt.savefig(os.path.join(output_dir, f'LISA_{nome_setor}_{ano}.png'), dpi=300, bbox_inches='tight')
    plt.show()

# Função para processar um único arquivo shapefile
def processar_arquivo(shapefile_path, anos_selecionados, output_dir):
    """
    Processa um único arquivo shapefile, calculando Moran e LISA para os anos selecionados.
    """
    # Extrai o nome do setor a partir do nome do arquivo, removendo a extensão .shp
    nome_setor = os.path.basename(shapefile_path).replace('transformado.shp', '').replace('_',' ')
    print(f'Processando setor: {nome_setor}')

    # Carregar o shapefile
    gdf = gpd.read_file(shapefile_path)

    # Iterar sobre os anos selecionados
    for ano in anos_selecionados:
        if ano in gdf.columns:
            print(f'Processando ano: {ano}')

            # Filtrar dados para o ano selecionado
            coluna = str(ano)

            # Chama a função para calcular Moran e LISA
            calcular_moran_lisa(gdf, coluna, output_dir, nome_setor, ano)
        else:
            print(f'Coluna para o ano {ano} não encontrada no shapefile.')

# Função para processar todos os arquivos em um diretório
def processar_todos_arquivos(input_dir, anos_selecionados, output_dir):
    """
    Processa todos os arquivos shapefiles em um diretório.
    """
    # Loop sobre os arquivos shapefiles no diretório
    for arquivo in os.listdir(input_dir):
        if arquivo.endswith('.shp'):
            shapefile_path = os.path.join(input_dir, arquivo)
            processar_arquivo(shapefile_path, anos_selecionados, output_dir)

# Definindo diretórios de entrada e saída, anos
input_dir = '/content/drive/MyDrive/PESQUISAS/ENSAIO_MUDANCAS_CLIMATICAS/SAIDA/06_ANALISE_ESPACIAL/SHP_TRANSFORMADO'
output_dir = '/content/drive/MyDrive/PESQUISAS/ENSAIO_MUDANCAS_CLIMATICAS/SAIDA/06_ANALISE_ESPACIAL/RESULTADOS_MORAN_LISA'

anos_selecionados = ['1997', '2003', '2010', '2015', '2023']

# Chama a função para processar todos os arquivos
processar_todos_arquivos(input_dir, anos_selecionados, output_dir)
```

```python
anos_selecionados = ['1997', '2003', '2010', '2015', '2023']


from textwrap import wrap

# Inicialização das variáveis
dados_totais = []
```

```python
import random
from matplotlib import cm
from matplotlib.colors import to_rgba
```

```python
# Função para ler e processar os dados dos arquivos
def processar_dados(files_list, anos_selecionados):
    dados_totais = []

    for file in files_list:
        df = gpd.read_file(file)  # Ler arquivo shapefile
        setor = os.path.basename(file).replace('.shp', '').replace('_', ' ')  # Nome do setor
        df = df[['Municipio'] + anos_selecionados]  # Filtrar colunas relevantes

        # Calcular o ranking dos 5 maiores emissores por ano
        for ano in anos_selecionados:
            ranking = df[['Municipio', ano]].sort_values(by=ano, ascending=False).head(5)
            ranking['Setor'] = setor  # Adicionar a coluna do setor
            dados_totais.append(ranking)  # Adicionar dados ao total

    return pd.concat(dados_totais)  # Concatenar todos os rankings

# Função para quebrar o texto nas caixinhas em várias linhas, apenas em espaços
def wrap_text(text, width):
    words = text.split(' ')  # Dividir o texto em palavras
    lines = []
    current_line = ""

    for word in words:
        # Se adicionar a palavra exceder o limite de largura, adiciona a linha atual e começa uma nova
        if len(current_line + ' ' + word) > width:
            lines.append(current_line)
            current_line = word
        else:
            if current_line:  # Se já tiver texto na linha, adiciona a palavra com espaço
                current_line += ' ' + word
            else:
                current_line = word
    lines.append(current_line)  # Adicionar a última linha
    return lines

# Função para gerar cores aleatórias
def gerar_cores_unicas(num_cores):
    cores = []
    random.seed(42)  # Para garantir a reprodutibilidade das cores

    while len(cores) < num_cores:
        # Gerar uma cor aleatória com formato rgba
        cor_aleatoria = to_rgba(np.random.rand(4))  # RGBA com aleatoriedade
        # Garantir que a cor não seja repetida
        if cor_aleatoria not in cores:
            cores.append(cor_aleatoria)

    return cores

# Função para criar o gráfico
def criar_grafico(ranking_completo, anos_selecionados):
    # Gerar cores aleatórias únicas para cada município
    municipios_unicos = ranking_completo['Municipio'].unique()
    num_municipios = len(municipios_unicos)
    municipio_colors = {municipio: cor for municipio, cor in zip(municipios_unicos, gerar_cores_unicas(num_municipios))}

    for setor in ranking_completo['Setor'].unique():
        setor_df = ranking_completo[ranking_completo['Setor'] == setor]

        # Criar a figura com tamanho ajustado para (12, 6)
        plt.figure(figsize=(12, 6))

        # Gráfico
        for i, ano in enumerate(anos_selecionados):
            ax = plt.gca()

            # Filtra o ranking para aquele ano
            ranking_ano = setor_df[setor_df['Ano'] == ano].sort_values(by='Emissao', ascending=False).head(5)

            # Desenhar as caixinhas (retângulos) para os municípios
            for j, (municipio, emissao) in enumerate(zip(ranking_ano['Municipio'], ranking_ano['Emissao'])):
                # Quebra o texto em linhas de acordo com a largura máxima
                wrapped_text = "\n".join(wrap_text(municipio, 10))  # Limita a largura, mas só quebra nas palavras

                x_pos = i * 2  # Posição X, maior espaçamento entre os anos
                y_pos = 4 - j  # Ajuste para a posição 1 no topo e 5 na base

                # Desenha a caixinha com a cor associada ao município
                ax.add_patch(plt.Rectangle((x_pos, y_pos), 1.2, 0.8, color=municipio_colors[municipio]))  # Caixinha

                # Centralizar o texto na caixinha
                ax.text(x_pos + 1.2 / 2, y_pos + 0.8 / 2, wrapped_text, ha='center', va='center', color='black', fontsize=10)

        # Ajustar os eixos
        plt.xlim(-0.5, len(anos_selecionados) * 2 - 0.5)  # Ajustando a largura do eixo X para que as caixinhas fiquem visíveis
        plt.ylim(-0.5, 5)  # Garantir que todas as posições fiquem dentro do gráfico

        # Customizar eixos
        ax.set_xticks(np.arange(len(anos_selecionados)) * 2 + 0.6)
        ax.set_xticklabels(anos_selecionados)
        ax.set_yticks(np.arange(5))
        ax.set_yticklabels([f'Posição {i+1}' for i in range(5)][::-1])  # Inverter as posições no eixo Y para a Posição 1 ficar no topo

        # Adicionar título
        plt.title(f'Ranking de Emissões por Município no Setor: {setor}')
        plt.xlabel('Ano')
        plt.ylabel('Posição')

        plt.tight_layout()
        plt.show()

# Processar os dados e criar o gráfico
ranking_completo = processar_dados(files_list, anos_selecionados)
ranking_completo = pd.melt(ranking_completo, id_vars=["Municipio", "Setor"], value_vars=anos_selecionados,
                           var_name="Ano", value_name="Emissao")
criar_grafico(ranking_completo, anos_selecionados)
```
